```python
import pandas as pd
import requests
import json

# URL of the JSON data
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_week.geojson"

response = requests.get(url)
data = response.json()
print(json.dumps(data, indent = 2, sort_keys=True))
```

```python
# Create list to store the coordinates[2] values
coordinates_2_list = []

# For loop to obtain all the depths
for feature in data['features']:
    coordinates_2_list.append(feature['geometry']['coordinates'][2])

# Create a DataFrame from the list
df = pd.DataFrame({'Depth': coordinates_2_list})
df
```
